```python
import requests
import json
import geopandas as gpd
from dotenv import load_dotenv
import os
from sqlalchemy import create_engine
load_dotenv('../.env')
```

```python
#Write the data to PostGIS
db_connection_url = os.getenv('DATABASE_URL')

# Create a SQLAlchemy engine
engine = create_engine(db_connection_url)
# Get the GeoDataFrame from PostGIS
sql = "SELECT * FROM ski_resorts_skiarea"
df = gpd.GeoDataFrame.from_postgis(sql, engine,geom_col="location")
# "C:\Users\andko\Downloads\ne_10m_admin_0_countries_usa\ne_10m_admin_0_countries_usa.shp"
df2 = gpd.read_file(r"C:\Users\andko\Downloads\ne_10m_admin_0_countries_usa\ne_10m_admin_0_countries_usa.shp") 
```

```python
north_america = df2[df2["SOVEREIGNT"].isin(['United States of America', 'Canada', 'Mexico'])]
north_america = north_america[['SOVEREIGNT','geometry']]
```

```python
north_america
```

```python
result = gpd.sjoin(df, north_america, how="inner", op="within")
result['lon'] = result.geometry.apply(lambda p: p.x)
result['lat'] = result.geometry.apply(lambda p: p.y)

```

```python
result.to_list
```

```python
from src.isochrone_params import Isochrone_params
ski_resorts = result.to_dict(orient='records')
for ski_resort in ski_resorts:      
    iso_params = Isochrone_params(240,15,ski_resort['lon'],ski_resort['lat'])
    contours = iso_params.generatecontours()
    for contour in contours:
        dict_params = iso_params.create_isochrone_request(contour_times=contour)
        json_params = json.dumps(dict_params)
        r = requests.get(f'http://localhost:8002/isochrone?json={json_params}')
        min_contour = min(contour)
        max_contour = max(contour)
        file = open(f"./data/contour_{ski_resort['skiarea_pk']}_{min_contour}_{max_contour}.geojson", "wb")
        file.write(r.content)
        file.close()
```

```python
import re
#Write the data to PostGIS
db_connection_url = os.getenv('DEV_DATABASE_URL')
# Create a SQLAlchemy engine
engine = create_engine(db_connection_url)


for filename in os.listdir('./data'):
    match = re.search("[0-9a-f]{8}-[0-9a-f]{4}-[0-9a-f]{4}-[0-9a-f]{4}-[0-9a-f]{12}", filename)
    ski_area_pk = match.group()
    contour_gdf = gpd.read_file(f'./data/{filename}')
    contour_gdf = contour_gdf[['contour', 'geometry']]
    contour_gdf = contour_gdf.dropna()
    contour_gdf['ski_area_id'] = ski_area_pk
    contour_gdf = contour_gdf.rename_geometry('geom')
    contour_gdf = contour_gdf.set_crs(epsg=4326, allow_override=True)
    # Write the GeoDataFrame to PostGIS
    contour_gdf.to_postgis(name='ski_resorts_ski_area_isochrone', con=engine, if_exists='append')
```
