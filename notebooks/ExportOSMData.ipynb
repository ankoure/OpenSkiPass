```python
import codecs
import osm2geojson

with codecs.open(r"C:\Users\andko\Downloads\data\input_lifts.osmjson", 'r', encoding='utf-8') as data:
    xml = data.read()

geojson = osm2geojson.json2geojson(xml, filter_used_refs=False, log_level='INFO')
# >> { "type": "FeatureCollection", "features": [ ... ] }
```

```python
import json 
str_geojson = json.dumps(geojson)
f = open("lifts.geojson", "a")
f.write(str_geojson)
f.close()
```

```python
import geopandas as gpd
gdf = gpd.read_file(r"C:\Users\andko\OneDrive\Documents\GitHub\OpenSkiPass\notebooks\lifts.geojson")
```

```python
import pandas as pd
import json
# Convert the JSON strings to Python dictionaries
gdf['tags'] = gdf['tags'].apply(json.loads)

# Normalize the JSON data
normalized_df = pd.json_normalize(gdf['tags'])

# Merge the normalized data back into the original GeoDataFrame
gdf = pd.concat([gdf.drop(columns=['tags']), normalized_df], axis=1)

# Display the result
print(gdf)
```
